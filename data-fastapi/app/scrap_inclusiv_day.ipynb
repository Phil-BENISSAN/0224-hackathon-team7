```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from datetime import datetime
from datetime import timedelta
from datetime import date
import warnings
from pprint import pprint
import json
from html.parser import HTMLParser
warnings.filterwarnings("ignore")
```

```python
try:
    driver = webdriver.Chrome()
except:
    driver = webdriver.Chrome(ChromeDriverManager(version="114.0.5735.90").install())
```

```python
driver.get('https://event.inclusivday.com/sessions?accessibilit=velotypie&page=4')
```

```python
time.sleep(3) 
accept_cookies =  driver.find_element(By.ID, "didomi-notice-agree-button")
driver.execute_script("arguments[0].click();", accept_cookies);
```

```python
time.sleep(3) 
conferences = []
confs = driver.find_elements(By.XPATH, '/html/body/div[3]/div/div/div/div[2]/div[1]/section/div/div/div[2]/div/div[4]/div/div/div[1]/div[2]/div/div/div/div/div/div/a')
for conf in confs :
        time.sleep(0.5) 
        conferences.append(conf.get_attribute('href')) 
```

```python
len(conferences)
```

```python
list_dico = []
for conference in conferences:
    driver.get(conference)
    time.sleep(1)
    try:
        title = driver.find_element(By.XPATH, '//*[@id="ct-e3b8f3ae-0b93-4168-9c2c-d10eacb329c7"]/div[1]/div[2]/div/section/h1/span/span').text
        description = driver.find_element(By.XPATH, '//*[@id="ct-087b2963-8760-4224-f27e-dba41318d14e"]/div[1]/div/section/section/section[1]/div/span/span').text
        interventant_1 = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div/div[2]/div[1]/section/div/div/div[1]/div/div[3]/div/div/div[1]/div/div[1]/a/div[2]/h4/span[1]/span').text
        interventant_3 = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div/div[2]/div[1]/section/div/div/div[1]/div/div[3]/div/div/div[1]/div/div[1]/a/div[2]/h4/span[3]/span').text
        dico = {'titre' :title , 'descriptif':description , 'intervenant': interventant_1 +" "+interventant_3 ,'langue':'francais', 'duree_conf' : '1h', 'duree_total' : '2h30'}
        list_dico.append(dico)
    except:
        pass
```

# mise en forme 


```python
liste_modilite = list(df_total['mobilite'].unique())
liste_modilite = liste_modilite[1:]
liste_modilite
```

```python
liste_type = list(df_total['type'].unique())
liste_type = liste_type[1:]
liste_type
```

```python
liste_theme = list(df_total['theme'].unique())
liste_theme = liste_theme[1:]
liste_theme
```

```python
import random
```

```python
print(random.choice(liste_modilite))
```

```python
for ind, val in enumerate(df_total['mobilite']):
    if pd.isna(df_total['mobilite'][ind]):
        val = random.choice(liste_modilite)
        df_total['mobilite'][ind] = val
```

```python
for ind, val in enumerate(df_total['type']):
    if pd.isna(df_total['type'][ind]):
        val = random.choice(liste_type)
        df_total['type'][ind] = val
    if df_total['type'][ind] == 'présentiel':
        df_total['type'][ind] = 'presentiel'
```

```python
for ind, val in enumerate(df_total['mobilite']):
    if df_total['type'][ind] == 'distance':
        df_total['mobilite'][ind] = 'France'
```

```python
for ind, val in enumerate(df_total['theme']):
    if pd.isna(df_total['theme'][ind]):
        val = random.choice(liste_theme)
        df_total['theme'][ind] = val
```

```python
# Origines ethniques et culturelles - origine
# Egalité des sexes - sexe, identité de genre
# Diversité des âges - age
# Orientations sexuelles - orientation sexuelle
# Handicap - handicap
# Etat de santé - état de santé
# Diversité des apparences physiques - apparence physique
```

```python
for ind, val in enumerate(df_total['theme']):
    if df_total['theme'][ind] =='origine':
        df_total['theme'][ind] = 'Origines ethniques et culturelles'
    elif df_total['theme'][ind] =='age':
        df_total['theme'][ind] = 'Diversité des ages'
    elif df_total['theme'][ind] =='orientation sexuelle':
        df_total['theme'][ind] = 'Orientations sexuelles'
    elif df_total['theme'][ind] == 'handicap':
        df_total['theme'][ind] = 'Handicap'
    elif df_total['theme'][ind] == 'etat de santé':
        df_total['theme'][ind] = 'Etat de santé'
    elif df_total['theme'][ind] == 'apparence physique':
        df_total['theme'][ind] = 'Diversité des apparences physiques'
    else:
        df_total['theme'][ind] = 'Egalité des sexes'
```
